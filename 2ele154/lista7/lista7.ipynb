### Exercício 4: Obtenha a Resposta uma Entrada Arbitrária da Função de Transferência:
$$G(s)=\frac{10}{s + 1}$$
1. sin(t + 30o)
2. 2cos(2t - 45o)
3. sin(t + 30o) - 2cos(2t - 45o)


```octave
% Função de Transferência do Sistema
G = tf([1], [1 1 1])

% Função das Entradas
t = 0:0.1:10;

e1 = sin(t + pi/6);
e2 = 2*cos(2*t - pi/4);
e3 = sin(t - pi/6) - 2*cos(2*t - pi/4);

% Obtendo a Função das Entradas
G1 = lsim(G, e1, t);
G2 = lsim(G, e2, t);
G3 = lsim(G, e3, t);

% Resposta do Tipo Rampa

plot(G1)
hold on;
plot(G2, 'k')
hold on;
plot(G3, 'r')
```
