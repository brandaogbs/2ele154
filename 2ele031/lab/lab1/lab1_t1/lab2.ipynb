```python
#03.py

import pygame

pygame.init()

# definicao da resolucao
display_width  = 800
display_height = 600

# definicao das cores (monokai)
gray  = (39,40,34)
orange = (253,151,31)
pink   = (249,38,114)
blue   = (102,217,239)
green  = (166,226,46)

# definicoes gerais
gameDisplay = pygame.display.set_mode((display_width, display_height))
pygame.display.set_caption("Projeto Integrador 1B!")
clock = pygame.time.Clock()

# definicao do player
carImg = pygame.image.load('figs/mario.png')

# desenha o carro na posicao x,y
def car(x, y):
    gameDisplay.blit(carImg, (x,y))

# posicao inicial
x = display_width*0.45
y = display_height*0.8

x_change = 0

# flag de controle
gameExit = False

while not gameExit:

    # tratamento dos eventos
    for event in pygame.event.get():
        # quit
        if event.type == pygame.QUIT:
            gameExit = True
            
        # botao foi pressionado
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                x_change = -5
            elif event.key == pygame.K_RIGHT:
                x_change = 5

        # botao foi solto
        if event.type == pygame.KEYUP:
            if event.key == pygame.K_LEFT or event.key == pygame.K_RIGHT:
                x_change = 0

    # background
    gameDisplay.fill(gray)

    # atualiza a posicao do carro
    x += x_change

    # desenha carro
    car(x,y)

    # atualiza a tela
    pygame.display.flip()
    clock.tick(60)

pygame.quit()
quit()
```

```python
#04.py

import time

import pygame


pygame.init()

# definicao da resolucao
displayWidth  = 800
displayHeight = 600

# definicao das cores (monokai)
gray   = (39,40,34)
orange = (253,151,31)
pink   = (249,38,114)
blue   = (102,217,239)
green  = (166,226,46)

# definicoes gerais
gameDisplay = pygame.display.set_mode((displayWidth, displayHeight))
pygame.display.set_caption("Projeto Integrador 1B!")
clock = pygame.time.Clock()

# definicao do carro
carImg = pygame.image.load('figs/mario.png')
carWidth = 100

# desenha o carro na posicao x,y
def car(x, y):
    gameDisplay.blit(carImg, (x,y))

# update no x do carro
def car_update(x, x_change):
    x += x_change

    if x > displayWidth - carWidth:
        x = displayWidth - carWidth
        print("Bateu direita")
    elif x < 0:
        x = 0
        print("Bateu esquerda")
    return x

# renderiza o texto
def text_objets(text, font, color):
    textSurface = font.render(text, True, color)
    return textSurface, textSurface.get_rect()

# apresenta texto na tela
def message_display(text, color):
    largeText = pygame.font.Font('freesansbold.ttf', 115)
    
    TextSurf, TextRect = text_objets(text, largeText, color)
    TextRect.center = (displayWidth/2, displayHeight/2)

    gameDisplay.blit(TextSurf, TextRect)
    pygame.display.update()
    
    time.sleep(2)

    game_loop()

def crash():
    message_display("Você Bateu!", pink)

def game_loop():

    # posicao inicial
    x = displayWidth*0.45
    y = displayHeight*0.8

    # armazena o deslocamento
    x_change = 0

    # flag de controle
    gameExit = False

    while not gameExit:

        # tratamento dos eventos
        for event in pygame.event.get():
            # quit
            if event.type == pygame.QUIT:
                pygame.quit()
                quit()                
            
            # botao foi pressionado
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_LEFT:
                    x_change = -5
                elif event.key == pygame.K_RIGHT:
                    x_change = 5

            # botao foi solto
            if event.type == pygame.KEYUP:
                if event.key == pygame.K_LEFT or event.key == pygame.K_RIGHT:
                    x_change = 0

        # background
        gameDisplay.fill(gray)

        # atualiza a posicao do carro
        x = car_update(x, x_change)

        # desenha carro
        car(x,y)

        # atualiza a tela
        pygame.display.flip()
        clock.tick(60)

game_loop()

pygame.quit()
quit()
```

```python
# 05.py
import time
import random

import pygame


pygame.init()

# definicao da resolucao
displayWidth  = 800
displayHeight = 600

# definicao das cores (monokai)
gray   = (39,40,34)
orange = (253,151,31)
pink   = (249,38,114)
blue   = (102,217,239)
green  = (166,226,46)

# cor dos blocos
blockColors = [orange, pink, blue, green]

# definicoes gerais
gameDisplay = pygame.display.set_mode((displayWidth, displayHeight))
pygame.display.set_caption("Example PyGame!")
clock = pygame.time.Clock()

# definicao do carro
carImg = pygame.image.load('figs/mario.png')
carWidth = 100

#
# Carro
#

# desenha o carro na posicao x,y
def car(x, y):
    gameDisplay.blit(carImg, (x,y))

# update no x do carro
def car_update(x, x_change):
    x += x_change

    if x > displayWidth - carWidth:
        x = displayWidth - carWidth
        crash()
    elif x < 0:
        x = 0
        crash()
    return x

#
# Blocos
#
def block(x, y, w, h, color):
    pygame.draw.rect(gameDisplay, color, [x, y, w, h])

def block_update(blockX, blockY, blockH, blockW, blockColor, blockSpeed):
    blockY += blockSpeed

    if blockY > displayHeight:
        blockY = 0 - blockH
        blockX = random.randrange(0, displayWidth)
        blockColor = random_color()

    return blockX, blockY, blockColor

#
#   Texto
#

# renderiza o texto
def text_objets(text, font, color):
    textSurface = font.render(text, True, color)
    return textSurface, textSurface.get_rect()

# apresenta texto na tela
def message_display(text, color):
    largeText = pygame.font.Font('freesansbold.ttf', 115)
    
    TextSurf, TextRect = text_objets(text, largeText, color)
    TextRect.center = (displayWidth/2, displayHeight/2)

    gameDisplay.blit(TextSurf, TextRect)
    pygame.display.update()
    
    time.sleep(2)

    game_loop()


#
# Aux
#

def random_color():
    return random.choice(blockColors)

def crash():
    message_display("Você Bateu!", pink)

def game_loop():

    # posicao inicial carro
    carX = displayWidth*0.45
    carY = displayHeight*0.8

    # posicao inicial bloco
    blockX = random.randrange(0, displayWidth)
    blockY = -displayHeight
    blockSpeed = 7
    blockW = 100
    blockH = 100
    blockColor = random_color()

    # armazena o deslocamento carro
    carXUpdate = 0

    # flag de controle
    gameExit = False

    while not gameExit:

        # tratamento dos eventos
        for event in pygame.event.get():
            # quit
            if event.type == pygame.QUIT:
                pygame.quit()
                quit()                
            
            # botao foi pressionado
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_LEFT:
                    carXUpdate = -5
                elif event.key == pygame.K_RIGHT:
                    carXUpdate = 5

            # botao foi solto
            if event.type == pygame.KEYUP:
                if event.key == pygame.K_LEFT or event.key == pygame.K_RIGHT:
                    carXUpdate = 0

        # background
        gameDisplay.fill(gray)

        # atualiza posicao do blocos
        blockX, blockY, blockColor =  block_update(blockX, blockY, blockH, blockW, blockColor, blockSpeed)
        block(blockX, blockY, blockW, blockH, blockColor)

        # atualiza a posicao do carro
        carX = car_update(carX, carXUpdate)
        car(carX,carY)

        # atualiza a tela
        pygame.display.flip()
        clock.tick(60)

game_loop()

pygame.quit()
quit()
```
